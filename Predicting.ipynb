```python
with open('vocabulary.pkl', 'rb') as fp:
    vocabulary = pickle.load(fp)
    print('vocabulary dictionary')
    print(vocabulary)
```

```python
nlp = spacy.load("en_core_web_sm")

def clean_text(text):
    doc = nlp(text)
    doc = [t.lemma_.lower() for t in doc if len(t)>2]
    doc = " ".join([char for char in doc if char not in string.punctuation])
    return doc
```

```python
new_model = tf.keras.models.load_model('saved_model/my_model')
```

```python
path = "/content/TEST_SET_CUSTOM_OCR_SCAN.txt"
if path[-4:] == ".txt":
    with open(path, encoding="utf8", errors='ignore') as f:
        lines = f.readlines()
elif path[-4:] == ".pkl":
    labels_ocr = pd.read_pickle(path, compression='infer', storage_options=None)
    # to array of strings

else:
  print("ERROR FILE READ")
```

```python
len(lines)
```

```python
cleaned_text = []
for i in range (len(lines)):
    cleaned_text.append(clean_text(lines[i]))
```

```python
len(cleaned_text)
```

```python
vectorizer2 = TfidfVectorizer(vocabulary=vocabulary)
tfidf_vectorizer_vectors2 = vectorizer2.fit_transform(cleaned_text)
docs2 = tfidf_vectorizer_vectors2.toarray()
```

```python
len(docs2)
```

```python
y_pred_ohe = new_model.predict(docs2) 
predictions = np.argmax(y_pred_ohe,axis=1)
```
